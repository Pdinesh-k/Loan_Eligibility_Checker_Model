```python
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
numeric_cols
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="mean")
imputer.fit(df[numeric_cols])
imputer.transform(df[numeric_cols])
```

```python
from sklearn.model_selection import train_test_split
train_val_df , test_df = train_test_split(df,test_size=0.2,random_state = 42)
train_df , val_df = train_test_split(train_val_df,test_size=0.25,random_state = 42)
```

```python
inputs = list(train_df.columns)[1:-1]
target = "Loan_Status"
train_inputs = train_df[inputs]
train_target = train_df[target]
```

```python
test_inputs = test_df[inputs]
test_target = test_df[target]
```

```python
val_inputs = val_df[inputs]
val_target = val_df[target]
```

```python
train_inputs[numeric_cols] = imputer.transform(train_df[numeric_cols])
test_inputs[numeric_cols] = imputer.transform(test_df[numeric_cols])
val_inputs[numeric_cols] = imputer.transform(val_df[numeric_cols])
```

```python
val_df[numeric_cols].isna().sum()
```

```python
categorical_cols
```

```python
imputer = SimpleImputer(strategy="most_frequent")
imputer.fit(df[categorical_cols])
train_inputs[categorical_cols] = imputer.transform(train_df[categorical_cols])
```

```python
test_inputs[categorical_cols] = imputer.transform(test_df[categorical_cols])
val_inputs[categorical_cols] = imputer.transform(val_df[categorical_cols])
```

```python
test_inputs[inputs].isna().sum()
```

```python
train_inputs[inputs].isna().sum()
```

```python
val_inputs[inputs].isna().sum()
```

Now we want to scale the inputs inorder to get accurate results

```python
df[numeric_cols]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[numeric_cols])
```

```python
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
```

```python
train_inputs
```

Now we are going to convert categorical columns to 0/1 inorder to predict

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False,handle_unknown="ignore")
enc.fit(df[categorical_cols])
```

```python
enc.categories_
```

```python
encoded_cols = list(enc.get_feature_names_out(categorical_cols))
encoded_cols
```

```python
train_inputs[encoded_cols] = enc.transform(train_inputs[categorical_cols])
test_inputs[encoded_cols] = enc.transform(test_inputs[categorical_cols])
val_inputs[encoded_cols] = enc.transform(val_inputs[categorical_cols])
```

```python
train_inputs
```

```python
numeric_cols
```

Okay it's time to train the model.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver="liblinear")
```

```python
model.fit(train_inputs[numeric_cols + encoded_cols],train_target)
```

```python
train_inputs[numeric_cols + encoded_cols]
```

```python
x_train = train_inputs[numeric_cols + encoded_cols]
x_val = val_inputs[numeric_cols + encoded_cols]
x_test = test_inputs[numeric_cols + encoded_cols]
```

Now it's turn to predict the training model

```python
train_pred = model.predict(x_train)
```

```python
train_target
```

Now we can check for the Accuracy of the model

```python
from sklearn.metrics import accuracy_score
accuracy_score(train_target,train_pred)
```

Now it's turn to predict the test and val model

```python
test_pred = model.predict(x_test)
val_pred = model.predict(x_val)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_target,test_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(val_target,val_pred)
```

We have high accuracy for the model training with 84% which which is a good for predicting model
